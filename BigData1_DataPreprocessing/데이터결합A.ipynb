```python
import pandas as pd

df1 = pd.DataFrame({
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"]},
        index=[0, 1, 2, 3])

df2 = pd.DataFrame({
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"]},
        index=[4, 5, 6, 7])

df3 = pd.DataFrame({
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"]},
        index=[8, 9, 10, 11])

df3
```

```python
result = pd.concat([df1, df2, df3])
```

```python
result
```

```python
# 열로 합치게 되면 인덱스를 남기기 위해 행이 12개가 생긴다
result = pd.concat([df1, df2, df3], axis = 1)
```

## Merge & Join

### Merge

```python
import pandas as pd

left = pd.DataFrame({
        "key1": ["K0", "K1", "K2", "K3"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"]})

right = pd.DataFrame({
        "key1": ["K0", "K1", "K2", "K3"],
        "key2": ["K0", "K0", "K0", "K0"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"]})

display(left)
display(right)
```

```python
result = pd.merge(left, right, on=["key1", "key2"])
display(result)
```

### parameter 변경

```python
pd.merge(left, right, on = ["key1", "key2"], how = "left")
```

```python
pd.merge(left, right, on = ["key1", "key2"], how = "right")
```

### indicator: 어떻게 Join됐는지 보여줌

```python
pd.merge(left, right, on = ["key1", "key2"], how = "outer", indicator = True)
```

### Join

```python
import pandas as pd

left = pd.DataFrame({
    "key1": [1, 2, 3, 4],
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"]})

right = pd.DataFrame({
    "C": ["C1", "C2", "C5"],
    "D": ["D1", "D2", "D5"]},
    index=[1, 2, 5])
```

```python
result1 = left.join(right, how="right")
result2 = right.join(left)

display(result1)
display(result2)
```

```python
result1 = left.join(right, how="inner")
result2 = left.join(right, how="outer")

display(result1)
display(result2)
```

### on

```python
result = left.join(right, on="key1")
result
```

```python
pd.merge(left, right, left_on="key1", right_index=True, how="left")
```

※ Join이 on으로 그리기 더 수월하다.
